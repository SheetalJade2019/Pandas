```python
# D
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style

# display first and last 5 rows

automobile=pd.read_csv('E:\\Python\\dataset\\Automobile_data.csv')
# print(automobile)

print('------FIRST-------')
print(automobile.head(5))
print()
print()
print('------FIRST-------')
print(automobile.tail(5))
```

```python
# display most expensive car name and price

print('--expensive_car--')
print()
expensive_car=automobile[['company','price']][automobile.price==automobile['price'].max()]
print(expensive_car)

# display less expensive car name and price

print()
print('--less_expensive_car--')
print()
less_expensive_car=automobile[['company','price']][automobile.price==automobile['price'].min()]
print(less_expensive_car)
```

```python
#Print All Toyota Cars details : groupby()
print('Print All Toyota Cars details')

company_group=automobile.groupby('company')
toyota_group=company_group.get_group('toyota')
print(toyota_group)
```

```python
#Count total cars per company
print('Count total cars per company')
print()
company_group=automobile.groupby('company')
for group in company_group.groups:
    print(group + '         ' + str(len(group)))
    
    
#print(automobile['company'].value_counts())
```

```python
#Find each company’s Higesht price car
print('Find each company’s Highest price car')
print()
company_group=automobile.groupby('company')
for name,group in company_group:
    print(name+'\t'+str(group['price'].max()))# +'\t'+ str(group['average-mileage'].mean()))
    print()
```

```python
# Find the average mileage of each car making company
print('Find the average mileage of each car making company')
print()
company_group=automobile.groupby('company')
for name,group in company_group:
    print(name+'\t'+ str(group['average-mileage'].mean()))
    print()
```

```python
# Sort all cars by Price column

automobile.sort_values(by=['price','company'],inplace=True, ascending=False)
print(automobile)
```

```python
# Merge two data frames and append the second data frame as a new column to the first data frame.

print('Merge two data frames')
print()
Car_Price = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'Price': [23845, 17995, 135925 , 71400]}
car_Horsepower = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'horsepower': [141, 80, 182 , 160]}

df_price=pd.DataFrame(Car_Price)
df_horsepower=pd.DataFrame(car_Horsepower)

merge=pd.merge(df_price,df_horsepower)
print(merge)
```

```python
#Create two data frames using two Dicts, Concatenate those two data frames and create a key for each data frame.
print('Create two data frames using two Dicts, Concatenate those two data frames and create a key for each data frame.')
print()
GermanCars = {'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 'Price': [23845, 171995, 135925 , 71400]}
japaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 'Price': [29995, 23600, 61500 , 58900]}

df_German=pd.DataFrame(GermanCars)
df_japanese=pd.DataFrame(japaneseCars)

concat=pd.concat([df_German,df_japanese],keys=['G','J'])
print(concat)
```
